```python
# label encoding using sklearn
import pandas as pd 
from sklearn import preprocessing

df = pd.read_csv('cat-in-the-dat-ii/train.csv')
df.ord_2 = df.ord_2.fillna('NONE')

# initialise LabelEncoder
lbl_enc = preprocessing.LabelEncoder()
df['ord_2v2'] = lbl_enc.fit_transform(df.ord_2.values)
df.ord_2v2.value_counts()
```
