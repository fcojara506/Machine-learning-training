```python
import matplotlib.pyplot as plt
import geopandas as gpd
from shapely.geometry import box
import os
import rasterio
import numpy as np
```

```python
AEZ = gpd.read_file('data/WorldCereal_AEZ.geojson')
AEZ_ids_full = AEZ.aez_id.unique()
AEZ.head(5)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(15, 15))
AEZ.plot(column = "aez_id",ax = ax)
```

```python
bbox_countries = {'UK':box(-10.5, 49.5, 2.5, 60),
                  'Chile':box(-75.5, -56, -66, -17.5),}

AEZ_ids_selected = list()

for bbox_country in bbox_countries:
    # Transform into a geopandas dataframe
    gdf_bbox = gpd.GeoDataFrame({'geometry': [bbox_countries[bbox_country]]}, crs=AEZ.crs)

    # Intersect AEZ dataframe with bbox
    AEZ_country = gpd.overlay(AEZ, gdf_bbox, how='intersection')

    # Extract unique aez_id values from the filtered dataframe
    AEZ_ids_selected.append(AEZ_country.aez_id.unique().tolist())
    
    #plot AEZ select zones 
    AEZ_country.plot(column = 'aez_id')
    
#unlist ids
AEZ_ids_selected = [item for sublist in AEZ_ids_selected for item in sublist][:1]

AEZ_ids_selected
```

```python
# full crop directories
crop_types_directories = {
        'Maize':'data/WorldCereal_2021_tc-maize-main_maize_classification',
        'Winter Cereals':'data/WorldCereal_2021_tc-wintercereals_wintercereals_classification',
        'Spring Cereals':'data/WorldCereal_2021_tc-springcereals_springcereals_classification'
}

#select dataset
crop_types_selected = {key: crop_types_directories[key] for key in ['Spring Cereals']}
```

```python
def read_raster_in_chunks(dataset):
    for ji, window in dataset.block_windows(1):
        band_window = dataset.read(window=window, indexes=1)
        yield band_window, ji
```

```python
demo = True
"""
Naming convention of the ZIP files is as follows:
WorldCereal_{year}_{season}_{product}_{classification|confidence}.zip

The actual AEZ-based GeoTIFF files inside each ZIP are named according to following convention:
{AEZ_id}_{season}_{product}_{startdate}_{enddate}_{classification|confidence}.tif
"""

if __name__ == '__main__':
    #select AEZ regions
    AEZ_id_list = AEZ_ids_selected if demo else AEZ_ids_full
    #select crop types
    crop_types = crop_types_selected if demo else crop_types_directories
    
    #loop through type of crop and region
    for crop_type, directory in crop_types.items():
        print(f"Processing crop type: {crop_type}")
        for aez_id in AEZ_id_list:
            
            print(f"Region: {aez_id}")
            dir_files = os.listdir(directory)
            
            # loop through every tif file
            for filename in dir_files:
                
                if filename.startswith(str(aez_id)) and filename.endswith('.tif'):
                    print(f"{filename}")
                    full_filename = os.path.join(directory,filename)
                    a = rasterio.open(full_filename)
                    fig, ax = plt.subplots(1, 1, figsize=(15, 15))
                    ax.imshow(a.read(1), cmap='pink')
                    break
                    
```
